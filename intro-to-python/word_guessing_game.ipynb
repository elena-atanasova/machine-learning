## Word Guessing Game

- Create an interactive game where the user tries to guess a hidden word letter by letter within a limited number of attempts.

```python
import random

# selects a random word
def select_random_word():
  words = ["python", "computer", "science", "hangman", "programming", "artificial", "intelligence"]
  return random.choice(words)

# display the word with underscores for unguessed letters
def display_word_progress(word, guessed_letters):
  display = ""
  for letter in word:
    if letter in guessed_letters:
      display += letter
    else:
      display += "_ "
  return display

# play the guessing game
def word_guessing_game():
  word_to_guess = select_random_word()
  guessed_letters = []
  attempts = 10

  print("Welcome to the Word Guessing Game!")
  print(display_word_progress(word_to_guess, guessed_letters))

  while attempts > 0:
    guess = input("Guess a letter: ").lower()

    # validate the input
    if len(guess) != 1 or not guess.isalpha():
      print("Please enter a single valid letter.")
      continue

    # check if the letter have already been guessed
    if guess in guessed_letters:
      print("You've already guessed that letter.")
      continue

    # add the valid letter to the guessed letters
    guessed_letters.append(guess)

    # check if the guessed letter is in the word
    if guess in word_to_guess:
      print("Good guess!")
    else:
      # if the letter is not in the word, decrement the 'attempts'
      attempts -= 1
      print(f"Incorrect! You have {attempts} attempts left.")

    # display the current progress of the word
    progress = display_word_progress(word_to_guess, guessed_letters)
    print(progress)

    # check if word is completed
    if "_" not in progress:
      print(f"Congratulations! You guessed the word: {word_to_guess}")
      break

    # stop game, if no attempts left
    if attempts == 0:
      print(f"Game over! The word was: {word_to_guess}")

# start the game
word_guessing_game()
```
