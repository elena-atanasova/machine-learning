## Data Generation and Plotting

- Implement two functions: `generate_data` and `plot_data`.
- The `generate_data` function takes the number of data points `n`, a `slope`, an `intercept`, and a `std` as input, and return a generated `x` and `y` arrays.
- The `x` array consists of uniformly distributed random points from (-10, 10). The `y` array is such that each element $y$ in the array follows a linear relationship with respect to $x$, with some error.

- The relationship between $x$ and $y$ is defined by the equation:

\begin{align*} y = mx + b + \epsilon \end{align*}

Where:
- $m$ is the `slope`, which determines how much $y$ changes for a unit change in $x$.
- $b$ is the `intercept`, which is the value of $y$ when $x$ is 0.
- $\epsilon$ follows a **normal** distribution over (0, `std`).
- $x$ follows a **uniform** distribution over (-10, 10).

$\epsilon$ represents the error term. In real-world data, observations rarely fall perfectly on a straight line due to various factors like measurement errors or unmodeled influences. This error term accounts for that variability. $\epsilon$ is a normally distributed error, meaning the errors are centered around zero and most errors are small, with larger errors being less frequent. The standard deviation `std` parameter controls the spread of this error; a higher standard deviation means the errors are more spread out, leading to more scattered data points around the line.

- The `plot_data` function should take the `x` and `y` arrays as input and plot them using the `matplotlib` library.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define variables for data generation
slope = 4
intercept = -2
std = 5
n = 50
```

```python
def generate_data(n, slope, intercept, std):
  x = np.random.uniform(-10, 10, n)
  e = np.random.normal(0, std, n)
  y = slope * x + intercept + e

  return x, y

def plot_data(x, y, title="Generated Data", xlabel="X values", ylabel="Y values"):
    plt.figure(figsize=(8, 5))
    plt.scatter(x, y, color='blue', alpha=0.7, label='Data points')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend()
    plt.grid(True)
    plt.show()

# generate and plot data
x_generated, y_generated = generate_data(n, slope, intercept, std)
plot_data(x_generated, y_generated)
```

### Generate data based on a different relationship:

$y = 2\pi \sin(x) + \text{error}$

```python
def generate_data_sine(n, std):
    x = np.random.uniform(-10, 10, n)
    e = np.random.normal(0, std, n)
    y = 2 * np.pi * np.sin(x)
    return x, y

# plot for sine data
x, y = generate_data_sine(100, std=2)
plot_data(x, y, title="Noisy Sine Wave", xlabel="X (radians)", ylabel="Y = 2sin(X)")
```
