## Dice Roll Simulation

- Use **for loops** to simulate rolling a dice multiple times and calculate the average result.
- Create a function `simulate_dice_rolls` that takes the number of rolls (`n`) as input and returns the average result of these rolls.
- Goal: Understand how to repeat actions in Python and explore concepts related to probability.

```python
import random
import matplotlib.pyplot as plt

def simulate_dice_rolls(n):
  roll_sum = 0
  average_roll = 0

  for x in range(0, n):
    dice_roll = random.randint(1, 6)
    roll_sum += dice_roll

  average_roll = roll_sum / n

  return average_roll

# test function
average_roll = simulate_dice_rolls(10)
print(f"The average result of 10 rolls is: {average_roll}")
```

## Modify the function to simulate rolling two dice and returning the average of their sum

```python
# simulate rolling two dice
# return the average of their sum
def simulate_two_dice_rolls(n):

  roll_sum = 0
  average_roll = 0

  for x in range(0, n):
    dice1 = random.randint(1, 6)
    dice2 = random.randint(1, 6)
    total = dice1 + dice2
    roll_sum += total

  average_roll = roll_sum / n

  return average_roll

average_roll = simulate_two_dice_rolls(5)
print(f"The average result of 10 rolls is: {average_roll}")
```

## Calculate the expected average result of rolling one die

- Does it match the result observed in the simulations, especially for large value of `n`?
- This is related to the **Law of Large Numbers**

### The Law of Large Numbers:
- As the number of trials `n` goes to infinity, the average gets closer to the true expected value.

Meaning:
- Small `n` → lots of randomness → averages vary
- Large `n` → randomness “cancels out” → average approaches the expected value

```python
# Expected average result of rolling a die
# (1 + 2 + 3 + 4 + 5 + 6) / 6 = 3.5

# use large value of n
average_roll = simulate_dice_rolls(100000)
print(average_roll)
```

```python
# observe results for different n
for n in [10, 100, 1000, 10000, 100000]:
    avg = simulate_dice_rolls(n)
    print(f"{n} rolls → average = {avg}")
```

## Plot the average result against the number of rolls `n` to visualise how the average converges as `n` increases?

- Tip: Use a list to store the averages for different `n` values and then use `matplotlib` to plot.

```python
def plot_avg_result_rolls():
  rolls = [10, 20, 50, 100, 2000, 5000, 10000]
  averages = []

  for n in rolls:
    avg = simulate_dice_rolls(n)
    averages.append(avg)

  # plot results
  plt.plot(rolls, averages, marker = 'o')
  plt.axhline(3.5, linestyle='--')  # expected value
  plt.xlabel("Number of rolls (n)")
  plt.ylabel("Average result")
  plt.title("Convergence of Dice Roll Average Toward Expected Value (3.5)")
  plt.grid(True)
  plt.show()


plot_avg_result_rolls()
```
