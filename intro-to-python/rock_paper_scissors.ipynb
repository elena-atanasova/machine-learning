**Rock, Paper, Scissors Game**

```python
import random
```

```python
# declare variables
user_score = 0
bot_score = 0

choices = ["rock", "paper", "scissors"]
```

```python
# best of 5 - first player to win 3 games, wins the entire game
while user_score < 3 and bot_score < 3:
  user_choice = input("Enter your choice (rock, paper, scissors) here: ").lower()
  if user_choice not in choices:
    print("Invalid choice! Please try again!")
    continue
  bot_choice = random.choice(choices)
  print(f"Bot chose: {bot_choice}")
  if user_choice == bot_choice:
    print("It's a tie!")
  elif (user_choice == "rock" and bot_choice == "scissors") or \
    (user_choice == "paper" and bot_choice == "rock") or \
    (user_choice == "scissors" and bot_choice == "paper"):
    print("You win!")
    user_score += 1
  else:
    print("You loose!")
    bot_score += 1
  print(f"Final score: You {user_score} - {bot_score} Bot")
```
